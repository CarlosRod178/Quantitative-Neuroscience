# Definitions

Probability distributions define the behavior of samples drawn from a population. Most of the time we want to use the sample to generalize our knowledge to the population. The process of attaining this knowledge is called statistical inference. To understand how to perform statistical inference, we need to understand some basics about probability distributions.

In the population, a random variable takes on certain outcomes, or values, with certain probabilities. Probabilities are numbers between 0 and 1 that are assigned to each possible outcome. If you have a representative sample from the population, you can compute a histogram of the number of occurrences of each outcome in the sample. For discrete data (see below), you can divide each number of occurrences by the total number of outcomes in the sample, which gives a probability distribution.

Working with probability distributions requires first answering three basic questions:

### 1) Do I have discrete or continuous data?


Probability distributions are typically divided into two separate categories, depending on the type of data you are working with.

The first data type is discrete: the outcomes can assume a finite set of particular values. In some cases, discrete data are *categorical*; that is, they take on discrete values, but the values have no natural ordering. An example of discrete data is coin flips (the values are heads or tails). 

In other cases, discrete data are *ordinal*; that is, they take on discrete values that have an intrinsic ordering. A classic example in neuroscience is the number of action potentials (a discrete, all-or-nothing event) evoked from a retinal ganglion cell by a flash of light ("5 quanta") versus no light ("0 quanta") in a given amount of time. There is an obvious ordering: e.g., 10 spikes are more than 2 spikes. This kind of data was initially reported [here](https://www.sciencedirect.com/science/article/abs/pii/0042698971900332?via%3Dihub) and is further described [here](https://www.annualreviews.org/doi/abs/10.1146/annurev.neuro.21.1.227):

### 2) What distribution best describes the data?

Once you have determined whether your data are discrete or continuous, you can start considering whether they likely come from a distribution that has a particular, well-understood form or not. See the inter-spike interval example, above. In many cases, these parametric forms follow naturally from assumptions about the nature of the process that generated the data; e.g., the binomial distribution is exactly what you get when you flip a coin a certain number of times.

An example in neuroscience that has had a strong influence on our understanding of the neural computations responsible for generating saccadic eye movements is based on the LATER model developed by Roger Carpenter and colleagues (initially reported here). The idea is that the typical reaction time (RT) distribution measured for simple saccade tasks (e.g., "look at the object when it appears") has a long rightward tail, such that the distribution of reciprocal RTs (i.e., the distribution of 1/RT) is Gaussian. The model accounts for this empirical observation via a simple "linear rise to a fixed bound" process, in which the rate of rise of the process is distributed as a Gaussian across saccades. Because RT is inversely related to the rate of rise of this process, the resulting RT distribution is an inverse Gausian, because rate x time = distance-to-bound (Fig 2 from here; example MATLAB code is here):

Neural correlates of this kind of rise-to-bound process were first described in the frontal eye field for simple visually guided saccades and since have been identified in a number of brain areas for numerous tasks.

### 3) What statistics or features best describe the distribution?

If your data are distributed according to a particular form, and that form can be described via a set of parameters, then you can take advantage of parametric statistics to describe and analyze the data. A common example is the use of the mean to describe the central trend of Gaussian-distributed data.

Otherwise, you can use non-parametric statistics that can work with any distribution but do not require assuming that it takes any particular form. A common example is the use of the median to describe the central trend of the data.

# Probability Distributions and Cumulative Probability Functions

The concept of a probability distribution is most easily understood for a categorical variable. In general, probabilities are numbers between 0 and 1, and a probability distribution describes the probabilities of each possible value of a particular categorical variable, such that all of the probabilities sum to 1. The cumulative distribution takes the same data and expresses them as the probability of obtaining a value equal to or less than each possible value. 

The relationship between probability and cumulative distributions has been used to understand how the brain encodes sensory information in an efficient way. For example, this paper proposes that a neuron that encodes the intensity of a stimulus should have responses that are most sensitive to (i.e., occur on the mid-point of the sigmoid-shaped function of response versus intensity, where changes in intensity induce the biggest changes in responses) intensity values that are most prevalent in the world.
