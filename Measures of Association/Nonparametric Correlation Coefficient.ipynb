# Definitions

Correlation is a measure of association between measured (random) variables. When you know that the variables are normally distributed, use the [Pearson's correlation coefficient](https://colab.research.google.com/drive/11kgk7FpLgbSlA4pjS4cCH1B5mYDUgj5b?usp=sharing). Otherwise, use non-parametric forms that compare paired values according to their ordinal scale: 1) the Kendall rank correlation coefficient, or 2) the Spearman rank correlation coefficient. They usually give very similar values and thus in most cases can be used interchangeably. When they give different answers, it probably means you should take a closer look at your data and figure out if something is wrong.

# Neuroscience example

From: Joshi et al (2016) [Relationships between Pupil Diameter and Neuronal Activity in the Locus Coeruleus, Colliculi, and Cingulate Cortex](https://www.sciencedirect.com/science/article/pii/S089662731501034X?via%3Dihub). 

Here we will examine in detail Figure 3 from that paper, which shows trial-by-trial associations between mean pupil diameter and spike rate measured in a number of brain areas.

It has been recognized for a long time that changes in pupil size occur not only in response to changes in ambient light levels, but also as a function of other ongoing cognitive operations; e.g., "deep emotions of pleasure as well as fear are commonly accompanied by pupillary dilation" ([Kuntz, 1929](https://catalog.hathitrust.org/Record/006153997);  "the pupilsâ€¦ register directly certain activities of the nervous system, including, but not restricted to, the effects of visual stimulation" ([Hess and Polt, 1964](https://science.sciencemag.org/content/143/3611/1190). As the number of studies that examine the conditions that give rise to these kinds of non-luminance-mediated changes in pupil size have grown, there has been a heightened interest in understanding the specific neural mechanisms that underlie them. One particular mechanism that has gained a great deal of attention is the locus coeruleus (LC), a small brainstem nucleus that is the source of the neuromodulator norepinephrine to most of the brain. Many articles that describe measures of pupil changes interpret those changes in terms of LC activity.

Interestingly, until recently it was possible to trace the basis for that claim to a single reference; e.g., 

- "Finally, neurophysiological recordings in the monkey have indicated that tonic changes in pupil diameter closely track the time course of LC activity, and show the same relationship to behavioral performance as tonic LC activity (Aston-Jones & Cohen, 2005; Rajkowski, Kubiak, & Aston-Jones, 1993)." ([Nieuwenhuis et al, 2011](https://pubmed.ncbi.nlm.nih.gov/20557480/)\).

- "As shown in Figure 7, pupil diameter correlates remarkably well with LC tonic activity in the monkey (Rajkowski et al. 1993)" ([Aston-Jones and Cohen, 2005](https://pubmed.ncbi.nlm.nih.gov/16022602/)\).

- "Rajkowski, Kubiak, and Aston-Jones (1993), for example, found a strong correlation in monkeys between baseline pupil diameter and tonic LC firing rate over the course of 90 min of performance in a target-detection task." ([Jepma et al, 2011](https://pubmed.ncbi.nlm.nih.gov/20666595/)\).

- "Rajkowski et al. (1993) found that baseline pupil diameter (recorded at the onset of each trial of the experiment) closely tracked LC tonic discharge frequency. Figure 1 provides an illustration of this strong correlation between LC discharge frequency and pupil diameter." ([Gilzenrat et al, 2010](https://pubmed.ncbi.nlm.nih.gov/20498349/)\).
